### using WikiText-103 dataset for training

```python
from pathlib import Path
import re
```

Merge

```python
DATA_FOLDER = Path.cwd().parent / "data"
TRAIN_FILE = DATA_FOLDER / "wiki.train.tokens"
VAL_FILE = DATA_FOLDER / "wiki.valid.tokens"
DATA_FILE = DATA_FOLDER / "corpus.txt"
```

```python
with open(TRAIN_FILE, "r", encoding="utf-8") as f:
    for _ in range(20):
        line = f.readline()
        print(repr(line))
```

```python
files = [TRAIN_FILE, VAL_FILE]

with open(DATA_FILE, "w", encoding="utf-8") as out:
    for file in files:
        with open(file, "r", encoding="utf-8") as f:
            out.write(f.read())
            out.write("\n")
```

```python
with open(DATA_FILE, "r", encoding="utf-8") as f:
    for _ in range(20):
        line = f.readline()
        print(repr(line))
```

Clean

```python
with open(DATA_FILE, "r", encoding="utf-8") as f:
    text = f.read()
```

```python
text = text.replace("<unk>", "")
text = text.replace("@-@", "-")
text = text.replace("@,@", ",")
text = text.replace("@.@", ".")
text = text.replace("@;@", ";")
text = text.replace("@:@", ":")
```

```python
text = re.sub(r"\n{3,}", "\n\n", text)
text = re.sub(r"[ \t]+", " ", text)
```

```python
CLEAN_FILE = DATA_FILE.parent / "data.txt"
```

```python
with open(CLEAN_FILE, "w", encoding="utf-8") as f:
    f.write(text)

print(f"Characters: {len(text):,}")
```
